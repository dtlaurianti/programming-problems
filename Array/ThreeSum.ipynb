Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

 

Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation: 
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
Notice that the order of the output and the order of the triplets does not matter.

Example 2:

Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.

Example 3:

Input: nums = [0,0,0]
Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.

 

Constraints:

    3 <= nums.length <= 3000
    -105 <= nums[i] <= 105

```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        res = set()
        p, n, z = [],[],[]
        # split numbers into positive, negative, and zeroes
        for num in nums:
            if num > 0:
                p.append(num)
            elif num < 0:
                n.append(num)
            else:
                z.append(num)
        # put unique numbers into O(1) lookup sets
        P = set(p)
        N = set(n)
        
        # if there is a zero, then check for pairs of numbers
        if len(z) > 0:
            for num in P:
                if -num in N:
                    res.add(tuple(sorted([num, -num, 0])))
        
        # if there are 3 zeroes, add that
        if len(z) > 2:
            res.add(tuple([0,0,0]))

        # look for positive complement of negative pairs
        for i in range(len(n)):
            for j in range(i+1, len(n)):
                if -(n[i] + n[j]) in P:
                    res.add(tuple(sorted([n[i],n[j], -(n[i] + n[j])])))
        
        # look for negative complement of positive pairs
        for i in range(len(p)):
            for j in range(i+1, len(p)):
                if -(p[i] + p[j]) in N:
                    res.add(tuple(sorted([p[i],p[j], -(p[i] + p[j])])))
        
        return res





        '''
        BRUTE FORCE METHOD
        triples = []
        nums = sorted(nums)
        # loop from the smallest number up
        for i in range(len(nums)):
            # loop from the biggest number down to the small number
            for j in range(len(nums)-1, i, -1):
                # loop over the numbers between them
                    for k in range(i+1, j):
                        if nums[i] + nums[j] + nums[k] == 0:
                            if [nums[i], nums[k], nums[j]] not in triples:
                                triples.append([nums[i], nums[k], nums[j]])
        return triples
        '''


```
