You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

 

Example 1:

Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]

Example 2:

Input: list1 = [], list2 = []
Output: []

Example 3:

Input: list1 = [], list2 = [0]
Output: [0]

 

Constraints:

    The number of nodes in both lists is in the range [0, 50].
    -100 <= Node.val <= 100
    Both list1 and list2 are sorted in non-decreasing order.



```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:

        # base cases
        if list1 is None:
            return list2
        if list2 is None:
            return list1

        #initalize the head node of the merged list
        merge = None
        if list1.val <= list2.val:
            merge = list1
            list1 = list1.next
        else:
            merge = list2
            list2 = list2.next

        head = merge

        # merge the lists
        while True:
            if list1 is None:
                merge.next = list2
                break
            elif list2 is None:
                merge.next = list1
                break
            if list1.val <= list2.val:
                merge.next = list1
                merge = merge.next
                list1 = list1.next
            else:
                merge.next = list2
                merge = merge.next
                list2 = list2.next

        return head

                
```
